```python
# Lowest Risk Major
gap = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
print(gap)
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
lowest_spread = clean_df.sort_values('Spread')
lowest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
# Highest Potential
highest_mid_90 = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_mid_90[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Highest risk/spread
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
# Highest Mid-Career Earning
highest_spread = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
highest_spread[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

```python
# Group By STEM, Business, HASS
clean_df.groupby('Group')['Undergraduate Major'].count().sort_values(ascending=False)
```

```python
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group')[['Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 10th Percentile Salary', 'Mid-Career 90th Percentile Salary', 'Spread']].mean().sort_values('Mid-Career Median Salary', ascending=False)
```
